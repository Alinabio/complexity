```python
#функция расчета лингвистической сложности
def count_ling_complexity(seq, len_window = 50, alphabet = 4):
    denominator = 4
    list_numerator = []
    #расчет знаменателя
    for i in range(2, len_window+1):
    
        if i**alphabet < len_window:
            denominator += i**alphabet
        else:
            denomonator_add = len_window - (i-1)
            denominator += denomonator_add


    for i in range(1, len_window):
        for c in range(len(seq)+1):
            mini_seq = seq[c:(c+i)]
            
            
            list_numerator.append(mini_seq)
    set_numerator = set(list_numerator)
    numerator = len(set_numerator)
    result = numerator/denominator
    return result

#расчет для конкретной последовательности
def linguistic_complexity(seq, len_window = 50, step = 100, alphabet = 4):
    with open(seq) as file:
        lines = file.readlines()[1:]
    
    sequence = lines[0]
    
    dict_data = dict()   
    i = 0
   
    for i in range(0, len(sequence)+1, step):
        ind = sequence[i:(i+len_window)]
        i += step
        
        dict_data[i] = count_ling_complexity(ind, len_window, alphabet)
    return dict_data
```

```python
import matplotlib.pyplot as plt

dictionary = linguistic_complexity('EPI_ISL_402124-ORF1ab.fasta', 50, 100, 4)
plt.plot(list(dictionary.keys()), dictionary.values(), color='b')
plt.show()
```
